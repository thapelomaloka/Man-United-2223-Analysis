# Collect Game results

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
#Opent the web page
page_url = "https://www.premierleague.com/results"
driver.get(page_url)
driver.maximize_window()

#Accept Cookies and close the advert
driver.find_element(By.ID, 'onetrust-accept-btn-handler').click()
driver.find_element(By.ID, 'advertClose').click()


# Wait for 10 seconds
time.sleep(1)

# Find the dropdown element for season
dropdown_season = driver.find_element(By.XPATH, '//div[@data-dropdown-current="compSeasons"]')

# Click on the dropdown to expand it
dropdown_season.click()

# Find the option element using XPath
option_season = driver.find_element(By.XPATH, '//*[@id="mainContent"]/div[3]/div[1]/section/div[3]/div[3]/ul/li[2]')

# Click on the option
option_season.click()

# Wait for 10 seconds
time.sleep(1)

# Find the dropdown element for club
dropdown_club = driver.find_element(By.XPATH, '//*[@id="mainContent"]/div[3]/div[1]/section/div[4]/div[2]')

# Click on the dropdown to expand it
dropdown_club.click()

# Find the option element using XPath
option_club = driver.find_element(By.XPATH, '//*[@id="mainContent"]/div[3]/div[1]/section/div[4]/div[3]/ul/li[15]')

# Click on the option
option_club.click()

time.sleep(1)

# Scroll down to the bottom of the page
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(5)  # Wait for the page to load
# Find the section with class "fixtures"
section = driver.find_element(By.CLASS_NAME, 'fixtures')

# Find all the list items within the section
list_items = section.find_elements(By.CLASS_NAME, 'match-fixture')

# Create an empty DataFrame to store the data
data2 = []
columns = ['Home Team', 'Away Team', 'Competition', 'Venue', 'Score']

# Iterate over the list items and extract the desired data
for item in list_items:
    home_team = item.get_attribute('data-home')
    away_team = item.get_attribute('data-away')
    competition = item.get_attribute('data-competition')
    venue = item.get_attribute('data-venue')
    score = item.find_element(By.CLASS_NAME, 'match-fixture__score').text.strip()

    # Append the data to the list
    data2.append([home_team, away_team, competition, venue, score])

# Create the DataFrame
PremierLeague  = pd.DataFrame(data2, columns=columns)


# Close the browser
driver.quit()
```

```python
# Concatenate Cup and League
Allgames = pd.concat([League,Cup])
Allgames = drop_column(Allgames,'Score')
Allgames.head()
```

# Number of Games Won, Lost and Drawn.
 

```python
def calculate_result(row):
    if row['Home Team'] == 'Man Utd':
        if row['Home'] > row['Away']:
            return 'W'
        elif row['Home'] < row['Away']:
            return 'L'
        else:
            return 'D'
    elif row['Away Team'] == 'Man Utd':
        if row['Away'] > row['Home']:
            return 'W'
        elif row['Away'] < row['Home']:
            return 'L'
        else:
            return 'D'
    else:
        return 'D'  # For all other cases where Man Utd is not involved, consider it as a draw

Allgames['Win/Loss/Draw'] = Allgames.apply(calculate_result, axis=1)
```

```python
# Restore the index to its default index
Allgames = Allgames.reset_index(drop=True)
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
#Opent the web page
page_url = "https://www.premierleague.com/clubs/12/Manchester-United/season-history"
driver.get(page_url)

# Maximize the window
driver.maximize_window()

#Accept Cookies and close the advert
driver.find_element(By.ID, 'onetrust-accept-btn-handler').click()
driver.find_element(By.ID, 'advertClose').click()


# Wait for 10 seconds
time.sleep(3)

# Initialize the DataFrame
data = pd.DataFrame(columns=['Season', 'LeaguePosition', 'Manager', 'TopGoalscorer'])

count = 1
while count <= 31:
    # Locate the elements using XPath
    season_element = driver.find_element(By.XPATH, '//*[@id="mainContent"]/div[2]/section[{}]/header/h2'.format(count))
    league_position_element = driver.find_element(By.XPATH, '//*[@id="mainContent"]/div[2]/section[{}]/div/div[1]/div/dl[1]/dd'.format(count))
    manager_element = driver.find_element(By.XPATH, '//*[@id="mainContent"]/div[2]/section[{}]/div/div[1]/div/dl[2]/dd/p'.format(count))
    top_goalscorer_element = driver.find_element(By.XPATH, '//*[@id="mainContent"]/div[2]/section[{}]/div/div[1]/div/dl[3]/dd/p'.format(count))
    
    # Extract the text from the elements
    season = season_element.text
    league_position = league_position_element.text
    manager = manager_element.text
    top_goalscorer = top_goalscorer_element.text
    
     # Create a temporary DataFrame for the current iteration
    temp_data = pd.DataFrame({'Season': [season], 'LeaguePosition': [league_position], 'Manager': [manager], 'TopGoalscorer': [top_goalscorer]})
    
    # Concatenate the temporary DataFrame with the main DataFrame
    data = pd.concat([data , temp_data], ignore_index=True)
    
    count += 1

print("Done")
```

```python
#View the 1st 5 rows
data.head()
```

```python
# Assuming you want to strip the characters '[' and ']' from the 'column_name' column
data['LeaguePosition'] = data['LeaguePosition'].str.replace('rd', '').str.replace('th', '').str.replace('nd', '')
data['Manager'] = data['Manager'].str.replace("Félix", '').str.replace('Mário dos Santos', '')
data['Season'] = data['Season'].str.replace("Champions", '')

data[['TopGoalscorer', 'Score']] = data['TopGoalscorer'].str.extract(r'(.*)\((\d+)\)')

# Convert the 'Score' column to integers
data['Score'] = data['Score'].astype(int)
data['LeaguePosition'] = data['LeaguePosition'].str.extract('(\d+)').astype(int)
```
